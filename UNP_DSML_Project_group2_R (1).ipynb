# Seperating Input Features and Target values

```python
# Features - All columns but last
x= df.iloc[: ,:-1].values
# Target - Last columns
y= df.iloc[: , -1].values
```

```python
# independent variable
x = df.drop(['CC_Strength'], axis =1)
# dependent variable
y = df['CC_Strength']
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
sc = StandardScaler()
Fit = sc.fit(x_train)
x_train = Fit.transform(x_train)
x_test = Fit.transform(x_test)
```

```python
####Applying Linear Regression
lr=LinearRegression()
fit=lr.fit(x_train,y_train)
y_pred_lr= lr.predict(x_test)
score = lr.score(x_test,y_test)
print('Predicted score is : {}'.format(score))
print('')
print('Mean_sqrd_Error_ =',mean_squared_error(y_test,y_pred_lr))
print('Mean_absolute_Error_ =',mean_absolute_error(y_test,y_pred_lr))
print('Cofficient_of_Determination_ =',r2_score(y_test,y_pred_lr))
rms = np.sqrt(mean_squared_error(y_test,y_pred_lr))
print('Root_mean_squared_Error_ = {}'.format(rms))
```

```python
    ####Applying Lasso Regression
ls= Lasso(alpha=0.4)
fit_ls = ls.fit(x_train,y_train)
y_pred_ls = ls.predict(x_test)
print('Score of Lasso Regression is:-',ls.score(x_test,y_test))
print('')
print('Mean_sqrd_Error =',mean_squared_error(y_test,ls.predict(x_test)))
print('Mean_absolute_Error =',mean_absolute_error(y_test,y_pred_ls))
print('Cofficient_of_Determination =',r2_score(y_test,y_pred_ls))
print('Root_mean_squared_Error =',np.sqrt(mean_squared_error(y_test,ls.predict(x_test))))
```

```python
####Applying Ridge Regression
rd= Ridge(alpha=0.4)
fit_rd = rd.fit(x_train,y_train)
y_pred_rd = rd.predict(x_test)
print('Score of Ridge Regression is:-',rd.score(x_test,y_test))
print('')
print('Mean_Sqrd_Error =',mean_squared_error(y_test,rd.predict(x_test)))
print('Mean_absolute_Error =',mean_absolute_error(y_test,y_pred_rd))
print('Cofficient_of_Determination =',r2_score(y_test,y_pred_rd))
print('Root_mean_squared Error =',np.sqrt(mean_squared_error(y_test,rd.predict(x_test))))
```

```python
####Applying Random Forest Regressor
rnd= RandomForestRegressor(ccp_alpha=0.4)
fit_rnd= rnd.fit(x_train,y_train)
y_pred_rnd = rnd.predict(x_test)
print('Score is:-',rnd.score(x_test,y_test))
print('')
print('Mean_Sqrd_Error =',mean_squared_error(y_test,rnd.predict(x_test)))
print('Mean_absolute_Error =',mean_absolute_error(y_test,y_pred_rnd))
print('Cofficient_of_Determination =',r2_score(y_test,y_pred_rnd))
print('Root_Mean_Squared Error=',np.sqrt(mean_squared_error(y_test,rnd.predict(x_test))))
```
